```python
PATH = 'hand_recognition/data/'
```

```python
img = cv.imread(PATH + 'yes/yes_0.jpg')
cv.imshow('Img',img)
k=cv.waitKey(0)
```

```python
import os
```

```python
data = []
output = []
for ind,i in enumerate(os.listdir(PATH)):
    for ind2,j in enumerate(os.listdir(PATH+i)):
        data.append(cv.imread(PATH+i+'/'+j))
        output.append(i)
```

```python
output[-1]
```

```python
len(output)
```

```python
import random
temp = list(zip(data, output))
random.shuffle(temp)
res1, res2 = zip(*temp)
# res1 and res2 come out as tuples, and so must be converted to lists.
data, output = list(res1), list(res2)
```

```python
classes={'yes':0,'no':1,'please':2,'good':3,'hello':4,'you':5,'nice_meet_you':6,'my_name':7,'good_morning':8,'how_you':9}
class_reverse = {0:'yes',1:'no',2:'please',3:'good',4:'hello',5:'you',6:'nice_meet_you',7:'my_name',8:'good_morning',9:'how_you'}
```

```python
for i in range(len(output)):
    output[i] = classes[output[i]]
```

```python
output
```

```python
for ind,i in enumerate(data):
    data[ind] = cv.resize(cv.cvtColor(i,cv.COLOR_BGR2GRAY),(32,32))
```
